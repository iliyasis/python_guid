## پایتون چیست؟

##### پایتون یک زبان برنامه‌نویسی **سطح بالا و بسیار محبوب** است که توسط **Guido van Rossum** ساخته شد و اولین نسخه‌ی آن در سال **1991** منتشر شد.

##### هدف طراحی پایتون این بود که زبانی ساده، خوانا و قدرتمند برای برنامه‌نویسی ایجاد شود.

##### امروزه پایتون یکی از پرکاربردترین زبان‌های برنامه‌نویسی در دنیا است.

## پایتون چه کارهایی می‌تواند انجام دهد؟

##### پایتون یک زبان **همه‌کاره** است و در حوزه‌های زیادی استفاده می‌شود.

### برخی از کاربردهای آن:

- توسعه وب (Backend)
- ساخت نرم‌افزارهای مختلف
- تحلیل داده و علم داده
- یادگیری ماشین و هوش مصنوعی
- اسکریپت‌نویسی و خودکارسازی کارها
- کار با پایگاه داده
- پردازش فایل‌ها و داده‌ها

## چرا پایتون؟

### چند دلیل مهم برای محبوبیت پایتون:

##### روی سیستم‌های مختلف اجرا می‌شود  (Windows, Linux, Mac)

##### **سینتکس ساده و خوانا** دارد و یادگیری آن آسان است.

##### برای انجام بسیاری از کارها **به کد کمتری** نسبت به بعضی زبان‌ها نیاز دارد.

##### یک زبان **Interpreted** است، یعنی برنامه‌ها سریع اجرا و تست می‌شوند.

- از چندین سبک برنامه‌نویسی پشتیبانی می‌کند:
  - Procedural
  - Object-Oriented
  - Functional

##### در بسیاری از زبان‌ها برای مشخص کردن بلوک کد از `{}` استفاده می‌شود.

##### اما در پایتون بلوک‌های کد با **تورفتگی (Indentation)** مشخص می‌شوند.

##### این ویژگی‌ها باعث می‌شود کد پایتون بسیار مرتب و خوانا باشد.
