## اجرای کد پایتون

### روش اول — محیط تعاملی (Command Line)

##### 🧠 این محیط به آن **Python Interactive Shell** گفته می‌شود.

</div>


```python
print("Hello, World!")
```

## اجرای فایل پایتون

##### مثال:

</div>


```python
# python myfile.py
```

## Indentation در پایتون

##### **Indentation** یعنی فاصله‌ای که در ابتدای یک خط کد قرار می‌گیرد.

##### در بسیاری از زبان‌های برنامه‌نویسی، تورفتگی فقط برای خوانایی کد است.

##### اما در **پایتون تورفتگی بخشی از Syntax زبان است.**

##### پایتون از تورفتگی برای مشخص کردن **بلاک‌های کد** استفاده می‌کند.

</div>


<div dir="rtl">

### مثال

##### در این مثال اگر شرط درست باشد، دستور چاپ اجرا می‌شود.

</div>


```python
if 5 > 2:
    print("salam")
```

<div dir="rtl">

#### ⚠️ **خطای رایج**

##### اگر تورفتگی را رعایت نکنیم، پایتون خطای Syntax می‌دهد.

</div>


```python
if 5 > 2:
print("salam")
```

<div dir="rtl">

## 📌 **نکته مهم**

##### تعداد فاصله‌ها را برنامه‌نویس تعیین می‌کند.

#### اما استاندارد رایج:

##### **۴ فاصله (4 spaces)**

##### مثال‌های زیر هر دو معتبر هستند.

</div>


```python
if 5 > 2:
 print("salam")
```

```python
if 5 > 2:
        print("salam")
```

<div dir="rtl">

## ⚠️ **خطای مهم**

##### در یک بلاک کد باید همه خطوط **تورفتگی یکسان** داشته باشند.

##### اگر تعداد فاصله‌ها متفاوت باشد، پایتون خطا می‌دهد.

</div>


```python
# سینتکس درست
if 5 > 2:
    print("salam")
    print("salam")
```

```python
# سینتکس غلط
if 5 > 2:
 print("salam")
        print("salam")
```

<div dir="rtl">

## متغیرها در پایتون

##### در پایتون متغیرها هنگام **اختصاص مقدار (assignment)** ساخته می‌شوند.

##### برخلاف بعضی زبان‌ها، در پایتون نیازی به تعریف نوع متغیر نداریم.

</div>


```python
x = 5
y = "Hello, World!"

print(x)
print(y)
```

<div dir="rtl">

## 🧠 **مفهوم مهم**

##### در پایتون نیازی به نوشتن نوع متغیر نیست.

##### مثلاً در زبان‌هایی مثل Java یا C باید بنویسیم:

`;int x = 5`

##### اما در پایتون فقط می‌نویسیم:

`x = 5`

##### پایتون نوع داده را **خودش تشخیص می‌دهد.**
###### توضیحات کامل تر در مورد انواع نوع داده را در جلسات بعد بررسی می کنیم

</div>

## کامنت‌ها (Comments)

##### کامنت‌ها برای **توضیح کد** استفاده می‌شوند.

##### کامنت‌ها توسط مفسر پایتون اجرا نمی‌شوند.

##### در پایتون کامنت با علامت `#` شروع می‌شود.

</div>


```python
# This is a comment

print("Hello, World!")
```

### برنامه‌ای بنویسید که:

- نام شما
- سن شما
- رشته تحصیلی شما

#### را چاپ کند.
### توضیحات کامل:
###### برنامه ای بنویسید که در آن سه متغیر ساخته شود و به ترتیب مقدار آنها را برابر با `نام` ، `سن` ، `رشته تصحیلی` قرار دهید؛ سپس هر سه آنها را چاپ کنید.

##### نکته : در کد خود یک کامت اضافه کنید که در آن توضیحاتی در مورد کد شما نوشته شده باشد.
