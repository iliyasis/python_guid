## متغیر چیست؟

##### **متغیر** یک ظرف (container) برای ذخیره‌ی مقادیر داده است.

##### در پایتون، متغیر **دقیقاً در لحظه‌ای که مقداردهی می‌شود** ساخته می‌شود.

##### برخلاف بعضی زبان‌ها، در پایتون نیازی به **اعلام (declare) کردن** متغیر قبل از استفاده نداریم.

</div>


```python
x = 5
y = "John"

print(x)
print(y)
```

<div dir="rtl">

## تغییر نوع متغیر

##### در پایتون، متغیرها نیاز به تعریف نوع ندارند.

##### همچنین می‌توان **نوع متغیر را پس از مقداردهی اولیه تغییر داد.**

</div>


```python
x = 4        # x یک عدد صحیح است
x = "Sally"  # حالا x یک رشته است

print(x)
```

<div dir="rtl">

## تبدیل نوع داده (Casting)

##### اگر بخواهیم **نوع داده‌ی یک متغیر را خودمان مشخص کنیم**، می‌توانیم از Casting استفاده کنیم.

##### این کار با توابع آماده‌ی پایتون انجام می‌شود:

- `str()` ← تبدیل به رشته
- `int()` ← تبدیل به عدد صحیح
- `float()` ← تبدیل به عدد اعشاری

</div>


```python
x = str(3)    # x برابر '3' می‌شود
y = int(3)    # y برابر 3 می‌شود
z = float(3)  # z برابر 3.0 می‌شود

print(x)
print(y)
print(z)
```

<div dir="rtl">

## گرفتن نوع متغیر با `type()`

##### با تابع `type()` می‌توانیم ببینیم **نوع داده‌ی یک متغیر چیست.**

</div>


```python
x = 5
y = "John"

print(type(x))
print(type(y))
```

<div dir="rtl">

## 🧠 نکته: کوتیشن تکی یا دوتایی؟

##### در پایتون، رشته‌ها (Strings) را می‌توان هم با **کوتیشن دوتایی** و هم با **کوتیشن تکی** تعریف کرد.

##### هر دو روش کاملاً معتبر و یکسان هستند.

</div>


```python
x = "John"
y = 'John'

print(x)
print(y)
print(x == y)  # هر دو یکسان هستند
```

<div dir="rtl">

## ⚠️ حساسیت به حروف کوچک و بزرگ (Case-Sensitive)

##### نام متغیرها در پایتون **به حروف کوچک و بزرگ حساس هستند.**

##### یعنی `a` و `A` دو متغیر **کاملاً متفاوت** هستند.

</div>


```python
a = 4
A = "Sally"

# a و A دو متغیر جداگانه هستند
print(a)
print(A)
```

<div dir="rtl">

## قوانین نام‌گذاری متغیرها

##### نام متغیر باید از این قوانین پیروی کند:

- باید با **حرف** یا **زیرخط (`_`)** شروع شود
- **نمی‌تواند** با عدد شروع شود
- فقط می‌تواند شامل حروف، اعداد، و زیرخط باشد (`A-z`, `0-9`, `_`)
- **نمی‌تواند** شامل فاصله (space) باشد
- **نمی‌تواند** کلمات کلیدی پایتون باشد (مثل `if`، `for`، `print` و ...)

</div>


```python
# نام‌های معتبر
myvar = "John"
my_var = "John"
_my_var = "John"
myVar = "John"
MYVAR = "John"
myvar2 = "John"

print(myvar)
print(my_var)
print(_my_var)
```

```python
# نام‌های غیرمعتبر — این‌ها خطا می‌دهند
# 2myvar = "John"   # شروع با عدد
# my-var = "John"   # استفاده از خط تیره
# my var = "John"   # استفاده از فاصله
```

<div dir="rtl">

## 🧠 سبک‌های نام‌گذاری متغیرهای چند کلمه‌ای

##### وقتی نام متغیر از چند کلمه تشکیل شده باشد، سه روش رایج وجود دارد:

| سبک | مثال | توضیح |
|-----|------|-------|
| **Camel Case** | `myVariableName` | کلمه اول کوچک، بقیه با حرف بزرگ |
| **Pascal Case** | `MyVariableName` | همه کلمات با حرف بزرگ |
| **Snake Case** | `my_variable_name` | کلمات با زیرخط جدا می‌شوند |

##### در پایتون **Snake Case** بیشتر استفاده می‌شود.

</div>


```python
# Camel Case
myVariableName = "John"

# Pascal Case
MyVariableName = "John"

# Snake Case (توصیه شده در پایتون)
my_variable_name = "John"

print(myVariableName)
print(MyVariableName)
print(my_variable_name)
```

<div dir="rtl">

## اختصاص چند مقدار به چند متغیر

##### در پایتون می‌توان **در یک خط** چند متغیر را مقداردهی کرد.

</div>


```python
x, y, z = "Orange", "Banana", "Cherry"

print(x)
print(y)
print(z)
```

<div dir="rtl">

## اختصاص یک مقدار به چند متغیر

##### همچنین می‌توان **یک مقدار** را **به چند متغیر** به‌صورت همزمان داد.

</div>


```python
x = y = z = "Orange"

print(x)
print(y)
print(z)
```

<div dir="rtl">

## Unpacking — باز کردن لیست در متغیرها

##### اگر مقادیر در یک **لیست** یا **تاپل** باشند، می‌توانیم آن‌ها را مستقیماً در متغیرها **باز (unpack) کنیم.**

</div>


```python
fruits = ["apple", "banana", "cherry"]

x, y, z = fruits

print(x)
print(y)
print(z)
```

<div dir="rtl">

## چاپ متغیرها با `print()`

##### تابع `print()` اغلب برای **خروجی گرفتن از متغیرها** استفاده می‌شود.

##### می‌توان چند متغیر را با **ویرگول (`,`)** از هم جدا کرد و همه را در یک خط چاپ کرد.

</div>


```python
x = "Python"
y = "is"
z = "awesome"

# چاپ با ویرگول — بهترین روش
print(x, y, z)
```

<div dir="rtl">

#### ⚠️ **خطای رایج**

##### نمی‌توان رشته و عدد را با `+` در `print()` ترکیب کرد.

</div>


```python
x = 5
y = "John"

# این خطا می‌دهد!
# print(x + y)

# روش درست — استفاده از ویرگول
print(x, y)
```

## متغیرهای Global

##### متغیرهایی که **خارج از یک تابع** تعریف می‌شوند، **Global** هستند.

##### متغیرهای Global از **هر جایی** در برنامه قابل دسترسی هستند — هم داخل توابع و هم خارج از آن‌ها.

</div>


```python
x = "awesome"  # متغیر Global

def my_function():
    print("Python is " + x)  # دسترسی به متغیر Global از داخل تابع

my_function()
```

<div dir="rtl">

## 📌 نکته مهم — متغیر Local در تابع

##### اگر داخل یک تابع متغیری با **همان نام** تعریف کنیم، متغیر **داخل تابع Local** است و متغیر Global را **تغییر نمی‌دهد.**

</div>


```python
x = "awesome"  # متغیر Global

def my_function():
    x = "fantastic"  # این یک متغیر Local جداگانه است
    print("Python is " + x)

my_function()

print("Python is " + x)  # متغیر Global تغییر نکرده
```

<div dir="rtl">

## کلمه‌کلیدی `global`

##### اگر بخواهیم **از داخل تابع** متغیر Global را تغییر دهیم، باید از کلمه‌کلیدی `global` استفاده کنیم.

</div>


```python
x = "awesome"

def my_function():
    global x          # اعلام می‌کنیم که x همان متغیر Global است
    x = "fantastic"   # حالا متغیر Global تغییر می‌کند

my_function()

print("Python is " + x)  # مقدار Global تغییر کرده
```
