## نوع داده چیست؟

##### در برنامه‌نویسی، **نوع داده (Data Type)** یک مفهوم بسیار مهم است.

##### متغیرها می‌توانند داده‌هایی از **انواع مختلف** را ذخیره کنند.

##### هر نوع داده **کارهای متفاوتی** می‌تواند انجام دهد.

##### برای مثال:

- با **اعداد** می‌توان عملیات ریاضی انجام داد
- با **رشته‌ها** می‌توان متن‌ها را به هم چسباند
- با **لیست‌ها** می‌توان مجموعه‌ای از مقادیر را ذخیره کرد

</div>


<div dir="rtl">

## انواع داده‌های پیش‌فرض در پایتون

##### پایتون به‌صورت پیش‌فرض انواع داده زیر را دارد:

| دسته‌بندی | نوع داده |
|-----------|----------|
| **متن (Text)** | `str` |
| **عددی (Numeric)** | `int`، `float`، `complex` |
| **دنباله (Sequence)** | `list`، `tuple`، `range` |
| **نگاشت (Mapping)** | `dict` |
| **مجموعه (Set)** | `set`، `frozenset` |
| **بولین (Boolean)** | `bool` |
| **باینری (Binary)** | `bytes`، `bytearray`، `memoryview` |
| **هیچ (None)** | `NoneType` |

##### در جلسات آینده هر کدام از این‌ها را به‌طور کامل بررسی می‌کنیم.

</div>


<div dir="rtl">

## گرفتن نوع داده با `type()`

##### با تابع `type()` می‌توانیم **نوع داده‌ی هر متغیر یا مقداری** را بفهمیم.

</div>


```python
x = 5
print(type(x))
```

```python
# بررسی نوع انواع مختلف
print(type("Hello"))   # str
print(type(20))        # int
print(type(20.5))      # float
print(type(True))      # bool
print(type(None))      # NoneType
```

<div dir="rtl">

## تعیین نوع داده هنگام مقداردهی

##### در پایتون، **نوع داده‌ی متغیر** دقیقاً در همان لحظه‌ای تعیین می‌شود که به آن مقدار می‌دهیم.

##### پایتون بر اساس **شکل و نوع مقدار** به‌طور خودکار نوع داده را تشخیص می‌دهد.

</div>


```python
x = "Hello World"              # str
print(x, "-->", type(x))

x = 20                         # int
print(x, "-->", type(x))

x = 20.5                       # float
print(x, "-->", type(x))

x = 1j                         # complex
print(x, "-->", type(x))

x = ["apple", "banana"]        # list
print(x, "-->", type(x))

x = ("apple", "banana")        # tuple
print(x, "-->", type(x))

x = range(6)                   # range
print(x, "-->", type(x))

x = {"name": "John", "age": 36} # dict
print(x, "-->", type(x))

x = {"apple", "banana"}        # set
print(x, "-->", type(x))

x = True                       # bool
print(x, "-->", type(x))

x = None                       # NoneType
print(x, "-->", type(x))
```

<div dir="rtl">

## 🧠 نکته مهم

##### پایتون از مقدار داده تشخیص می‌دهد که نوع آن چیست:

- عدد بدون نقطه → `int`
- عدد با نقطه اعشاری → `float`
- متن داخل کوتیشن → `str`
- مقادیر داخل `[]` → `list`
- مقادیر داخل `()` → `tuple`
- مقادیر داخل `{}` با کلید → `dict`
- مقادیر داخل `{}` بدون کلید → `set`
- `True` یا `False` → `bool`

</div>


<div dir="rtl">

## تعیین نوع داده با توابع سازنده (Constructor)

##### اگر بخواهیم **نوع داده را خودمان به‌صراحت مشخص کنیم**، می‌توانیم از **توابع سازنده** استفاده کنیم.

##### این کار همان **Casting** است که در جلسه قبل یاد گرفتیم، ولی برای تمام انواع داده.

</div>


```python
x = str("Hello World")    # str
print(x, "-->", type(x))

x = int(20)               # int
print(x, "-->", type(x))

x = float(20.5)           # float
print(x, "-->", type(x))
```

```python
x = list(("apple", "banana", "cherry"))   # list
print(x, "-->", type(x))

x = tuple(("apple", "banana", "cherry"))  # tuple
print(x, "-->", type(x))

x = dict(name="John", age=36)             # dict
print(x, "-->", type(x))

x = set(("apple", "banana", "cherry"))    # set
print(x, "-->", type(x))

x = bool(5)                               # bool
print(x, "-->", type(x))
```

##### در این جلسه فقط با کلیات Data Types آشنا شدیم.

##### در جلسات بعدی هر کدام از انواع مهم را به‌طور کامل بررسی می‌کنیم:

- **Numbers** → `int`، `float`، `complex`
- **Strings** → `str` و متدهای آن

- **Lists** → `list` و عملیات روی آن

- **Sets** → `set`
