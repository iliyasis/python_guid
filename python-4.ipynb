<div dir="rtl">

## سه نوع عددی در پایتون

##### پایتون سه نوع عددی پیش‌فرض دارد:

- `int` ← عدد صحیح
- `float` ← عدد اعشاری
- `complex` ← عدد مختلط

##### هر سه نوع دقیقاً مثل بقیه‌ی متغیرها، **هنگام مقداردهی** ساخته می‌شوند.

</div>


```python
x = 1      # int
y = 2.8    # float
z = 1j     # complex

print(type(x))
print(type(y))
print(type(z))
```

<div dir="rtl">

## عدد صحیح — `int`

##### `int` یک عدد **صحیح** است — مثبت یا منفی — **بدون هیچ قسمت اعشاری**.

##### در پایتون، اعداد صحیح **طول نامحدود** دارند.

##### یعنی می‌توانند به هر اندازه‌ای بزرگ باشند!

</div>


```python
x = 1
y = 35656222554887711
z = -3255522

print(type(x))
print(type(y))
print(type(z))
```

<div dir="rtl">

## عدد اعشاری — `float`

##### `float` یا **"عدد ممیز شناور"** یک عدد مثبت یا منفی است که **یک یا چند رقم اعشاری** دارد.

</div>


```python
x = 1.10
y = 1.0
z = -35.59

print(type(x))
print(type(y))
print(type(z))
```

<div dir="rtl">

## 🧠 نکته — نماد علمی در `float`

##### اعداد اعشاری می‌توانند به **نماد علمی** هم نوشته شوند.

##### از حرف `e` یا `E` برای نشان دادن **توان ۱۰** استفاده می‌شود.

</div>


```python
x = 35e3    # یعنی 35 × 10³ = 35000.0
y = 12E4    # یعنی 12 × 10⁴ = 120000.0
z = -87.7e100

print(x)
print(y)
print(z)
print(type(x))
```

<div dir="rtl">

## عدد مختلط — `complex`

##### اعداد مختلط با **پسوند `j`** نوشته می‌شوند.

##### این اعداد از دو بخش تشکیل شده‌اند:
- **بخش حقیقی (real)**
- **بخش موهومی (imaginary)** که با `j` مشخص می‌شود

##### اعداد مختلط بیشتر در **محاسبات علمی و ریاضی پیشرفته** استفاده می‌شوند.

</div>


```python
x = 3+5j
y = 5j
z = -5j

print(x)
print(y)
print(z)
print(type(x))
```

<div dir="rtl">

## تبدیل انواع عددی به هم

##### می‌توان با توابع `int()`، `float()`، و `complex()` اعداد را **از یک نوع به نوع دیگر** تبدیل کرد.

##### این کار را **Type Conversion** یا **Type Casting** می‌نامند.

</div>


```python
x = 1      # int
y = 2.8    # float
z = 1j     # complex

# تبدیل int به float
a = float(x)

# تبدیل float به int
b = int(y)

# تبدیل int به complex
c = complex(x)

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

<div dir="rtl">

## ⚠️ نکته مهم

##### اعداد مختلط (`complex`) را **نمی‌توان** به `int` یا `float` تبدیل کرد.

</div>


```python
z = 1j

# این خطا می‌دهد!
# int(z)
# float(z)
```

## اعداد تصادفی — ماژول `random`

##### پایتون تابع آماده‌ای برای تولید عدد تصادفی **ندارد**.

##### اما یک **ماژول داخلی** به نام `random` دارد که این کار را می‌کند.

##### برای استفاده از آن باید ابتدا ماژول را **import** کنیم.

##### 📌 در جلسات بعدی ماژول‌ها را کامل‌تر یاد می‌گیریم.

</div>


```python
import random

# یک عدد تصادفی بین ۱ تا ۹ چاپ می‌کند
print(random.randrange(1, 10))
```

```python
import random

# هر بار اجرا کنید، عدد متفاوتی می‌دهد!
for i in range(5):
    print(random.randrange(1, 100))
```
