# جلسه ۵ — تبدیل نوع داده (Casting) در پایتون

##### در این جلسه با مفهوم **Casting** و تبدیل انواع داده به یکدیگر آشنا می‌شویم.

#### موضوعات این جلسه:

- Casting چیست؟
- تبدیل به عدد صحیح با `int()`
- تبدیل به عدد اعشاری با `float()`
- تبدیل به رشته با `str()`
- نکات مهم و خطاهای رایج
- تمرین عملی

</div>


<div dir="rtl">

## Casting چیست؟

##### گاهی اوقات می‌خواهیم **نوع داده‌ی یک متغیر را خودمان مشخص کنیم.**

##### این کار را **Casting** می‌نامند.

##### پایتون یک زبان **شی‌گرا** است و برای تعریف انواع داده از **کلاس** استفاده می‌کند.

##### به همین دلیل، Casting در پایتون با **توابع سازنده (Constructor Functions)** انجام می‌شود:

- `int()` ← ساخت عدد صحیح
- `float()` ← ساخت عدد اعشاری
- `str()` ← ساخت رشته

</div>


<div dir="rtl">

## تبدیل به عدد صحیح — `int()`

##### تابع `int()` می‌تواند از سه منبع مختلف یک عدد صحیح بسازد:

- از یک **عدد صحیح** دیگر
- از یک **عدد اعشاری** (قسمت اعشاری حذف می‌شود، گرد **نمی‌شود**)
- از یک **رشته** (فقط اگر رشته یک عدد صحیح باشد)

</div>


```python
# تبدیل از عدد صحیح
x = int(1)

# تبدیل از عدد اعشاری (قسمت اعشاری حذف می‌شود)
y = int(2.8)

# تبدیل از رشته
z = int("3")

print(x)
print(y)
print(z)

print(type(x))
print(type(y))
print(type(z))
```

<div dir="rtl">

## ⚠️ نکته مهم — `int()` گرد نمی‌کند!

##### وقتی `float` را به `int` تبدیل می‌کنیم، قسمت اعشاری فقط **حذف** می‌شود.

##### پایتون عدد را **گرد نمی‌کند.**

</div>


```python
print(int(2.2))   # نتیجه: 2 (نه 2)
print(int(2.9))   # نتیجه: 2 (نه 3 !)
print(int(-2.9))  # نتیجه: -2 (نه -3)
```

<div dir="rtl">

## تبدیل به عدد اعشاری — `float()`

##### تابع `float()` می‌تواند از سه منبع مختلف یک عدد اعشاری بسازد:

- از یک **عدد صحیح**
- از یک **عدد اعشاری** دیگر
- از یک **رشته** (اگر رشته یک عدد صحیح یا اعشاری باشد)

</div>


```python
# تبدیل از عدد صحیح
x = float(1)

# تبدیل از عدد اعشاری
y = float(2.8)

# تبدیل از رشته (عدد صحیح)
z = float("3")

# تبدیل از رشته (عدد اعشاری)
w = float("4.2")

print(x)
print(y)
print(z)
print(w)

print(type(x))
```

<div dir="rtl">

## تبدیل به رشته — `str()`

##### تابع `str()` می‌تواند انواع مختلف داده را به **رشته** تبدیل کند:

- از یک **عدد صحیح**
- از یک **عدد اعشاری**
- از یک **رشته** دیگر

</div>


```python
# تبدیل از عدد صحیح
x = str(2)

# تبدیل از عدد اعشاری
y = str(3.0)

# تبدیل از رشته
z = str("s1")

print(x)
print(y)
print(z)

print(type(x))
print(type(y))
print(type(z))
```

<div dir="rtl">

## 🧠 کاربرد واقعی `str()` — ترکیب عدد و متن

##### یکی از کاربردهای مهم `str()` وقتی است که بخواهیم **عدد و متن را کنار هم چاپ کنیم.**

##### همان‌طور که در جلسه‌ی قبل دیدیم، نمی‌توان عدد و رشته را مستقیماً با `+` ترکیب کرد.

</div>


```python
age = 25

# این خطا می‌دهد!
# print("سن من: " + age)

# روش درست — تبدیل عدد به رشته
print("سن من: " + str(age))
```

<div dir="rtl">

## ⚠️ خطاهای رایج در Casting

##### نمی‌توان هر رشته‌ای را به عدد تبدیل کرد.

##### اگر رشته حاوی **حروف** باشد، تبدیل به `int` یا `float` خطا می‌دهد.

</div>


```python
# این‌ها کار می‌کنند:
print(int("10"))      # ✅
print(float("3.14"))  # ✅

# این‌ها خطا می‌دهند:
# print(int("3.5"))   # ❌ رشته‌ی اعشاری به int تبدیل نمی‌شود
# print(int("abc"))   # ❌ حروف به int تبدیل نمی‌شوند
# print(float("hi"))  # ❌ حروف به float تبدیل نمی‌شوند
```

<div dir="rtl">

## 📌 جدول خلاصه — چه تبدیلی ممکن است؟

| از | به `int()` | به `float()` | به `str()` |
|----|-----------|------------|----------|
| `int` | ✅ | ✅ | ✅ |
| `float` | ✅ (اعشار حذف می‌شود) | ✅ | ✅ |
| `str` (عدد صحیح) | ✅ | ✅ | ✅ |
| `str` (عدد اعشاری) | ❌ | ✅ | ✅ |
| `str` (حروف) | ❌ | ❌ | ✅ |
| `complex` | ❌ | ❌ | ✅ |

</div>
